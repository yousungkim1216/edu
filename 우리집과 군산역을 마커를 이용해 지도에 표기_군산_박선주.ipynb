# 우리집 지도

```python
# 우리집을 지도에 표시 - 군산 오션클래스

import folium

# 위도
latitude = 35.993629455111
# 경도
longitude = 126.753808243
m = folium.Map(location=[latitude, longitude],
               zoom_start=17, 
               width=750, 
               height=500
              )

m
```

# 군산역

```python
# 군산역 입구에 마커로 표시

folium.Marker([latitude, longitude],
              popup="군산역",
              tooltip="군산역 입구").add_to(m)
m
```

```python
#  군산역 입구에 마커스타일 변경 - tooltip

folium.Marker([latitude, longitude],
              popup="군산역",
              tooltip="군산역 입구", 
              icon=folium.Icon('pink', icon='star'), # 색, 모양변경
             ).add_to(m)
m
```

```python
#  군산역 주변을 반경 50 원형으로 강조

folium.CircleMarker([latitude, longitude],
                    color='tomato',
                    radius = 50, 
                    tooltip='군산역').add_to(m)
m
```
