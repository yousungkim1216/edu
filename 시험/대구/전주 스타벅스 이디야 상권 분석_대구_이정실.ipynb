```python
import pandas as pd

store = pd.read_csv('소상공인시장진흥공단_상가(상권)정보_전북_202206.csv', encoding='cp949' , low_memory=False)
```

```python
#shape : 행과 열의 개수
store.shape
```

```python
#열 레이블 (제목)
store.columns
```

```python
store.head(2)
```

```python
store = store[['시군구명','상권업종대분류명','상권업종중분류명','위도','경도','상호명','지점명']]
store.head(2)
```

```python
#결측치 확인(값이 없는 것)
store.isnull().sum()
```

```python
#스토어 안에서 시군구명의 전주시만 뽑아 오겠다. jeonju(전주로 이름 정하기)
#시군구명이 전주시 인 것을 검색 (contains)
jeonju=store[store['시군구명'].str.contains('전주시')]
jeonju
```

```python
#커피점/카페 추출
coffee=jeonju[jeonju['상권업종중분류명']=='커피점/카페']
coffee.head(2)
print('전주 지역 카페수 : ',len(coffee))
```

```python
#스타벅스 추출 (sb:스타벅스)
sb=coffee[coffee['상호명'].str.contains('스타벅스')]
print('스타벅스 수 : ', len(sb))
sb.head(2)
```

```python
#이디야 추출
edy=coffee[coffee['상호명'].str.contains('이디야')]
print('이디야 수 : ', len(edy))
edy.head(2)
```

```python
#구별 스타벅스 수
sb['시군구명'].value_counts()
```

```python
#구별 이디야 수
edy['시군구명'].value_counts()
```

```python
#스타벅스 이디야 데이터 프레임(시리즈) 데이터 합치는 작업
sb1=pd.Series(sb['시군구명'].value_counts(), name='스타벅스')
edy1=pd.Series(edy['시군구명'].value_counts(), name='이디야')
#concat:합침, axis 열방향, 1 한 열로 
sb_edy=pd.concat([sb1, edy1],axis=1)
sb_edy
```

```python
from folium.plugins import MarkerCluster
import folium

#전체 이디야 위치의 평균 mean()
m = folium.Map(location = [edy['위도'].mean(),
              edy['경도'].mean()],
              zoom_start = 12)

marker_cluster = MarkerCluster().add_to(m)

for lat, long, name in zip(sb['위도'], sb['경도'], sb['지점명']):
    folium.Marker([lat, long], icon = folium.Icon(icon='star', color='green'),
                 tooltip=name,
                 popup = f'<pre>스타벅스<br>{name}</pre>').add_to(marker_cluster)
m
```
