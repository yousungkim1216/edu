```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dg_metro = pd.read_csv('dg_m_20220731.csv', encoding='cp949')
dg_metro.head()
```

```python
dg_metro_7 = dg_metro[ dg_metro['월']==7]
dg_metro_7
```

```python
dg_metro_one = dg_metro_7 = dg_metro_7[dg_metro_7['역번호']<2000]

dg_metro_7
```

```python
dg_metro_on = dg_metro_one.loc[::2,]
#dg_metro_on
```

```python
dg_metro_off = dg_metro_one.loc[1::2,]
```

```python
list(set(dg_metro['월']))
```

```python
print(sorted(list(set(dg_metro['역명']))))
print(len(list(set(dg_metro['역명']))))
```

```python
dg_metro_on = dg_metro_on.groupby('역명').mean().reset_index()
dg_metro_on = dg_metro_on.drop(columns='일').set_index('역명')
dg_metro_on = dg_metro_on.mean(axis=1).sort_values(ascending=False)
dg_metro_on
```

```python
plt.figure(figsize=(20,10))
dg_metro_on.plot(kind=('bar'))
plt.title('대구도시철도 1호선 202207 승차분석')
plt.ylabel('승객수')
plt.show()
```
