```python
import folium

# 위도
latitude = 37.553664
# 경도
longitude = 126.9653595

m2 = folium.Map(
    location=[latitude, longitude],
    zoom_start=15
)

m2

from folium.plugins import MarkerCluster

marker_cluster = MarkerCluster().add_to(m2)

for lat, lng, name in zip(df['위도'], df['경도'], df['단속지점명']):
   folium.Marker([lat,lng], icon = folium.Icon(color='red'), popup = f'<pre>{name}</pre>').add_to(marker_cluster)
     
m2
```

```python
import glob
files = glob.glob('*불법주정차*')
files

gangnam = files[0]

pd.read_csv(gangnam, encoding='cp949')
pd.read_csv(gangnam, encoding='cp949').dtypes
```

```python
df_list = []
for file in files:
    try:
        raw = pd.read_csv(file, encoding='cp949')
        df_list.append(raw)
    except:
        raw = pd.read_csv(file, encoding='utf-8')
        df_list.append(raw)
        
df_list

```

```python
df = pd.concat(df_list)    # 976 rows × 6 columns

df
```

```python
df.reset_index(drop=True, inplace=True)  # 다시 드랍을 시키고..

df
```
