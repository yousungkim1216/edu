```python
# 지하철 공공데이터 분석(2022.02~06월)
import pandas as pd
import glob
```

```python
files = glob.glob('CARD*')
files
```

```python
f2 = files[0]
pd.read_csv(f2, encoding='cp949')
```

```python
pd.read_csv(f2, encoding='cp949').dtypes
```

```python
df_list = []
for file in files:
    try:
        raw = pd.read_csv(file, encoding='cp949', parse_dates=[0])
        df_list.append(raw)
    except:
        raw = pd.read_csv(file, encoding='utf-8', parse_dates=[0])
        df_list.append(raw)
        
#df_list
```

```python
df = pd.concat(df_list)
df
```

```python
df.reset_index(drop=True, inplace=True)
df
df.info()
```

```python
df.isnull().sum(0)
```

```python
df = df.drop('등록일자', axis=1)
df.insert(1, 'year', df['사용일자'].dt.year)
df.insert(2, 'month', df['사용일자'].dt.month)
df.insert(3, 'day', df['사용일자'].dt.day)
df
```

```python
df['total'] = df['승차총승객수'] + df['하차총승객수']
#df
```

```python
df.sort_values(by='total', ascending=False).head(10)
```

```python
# 역이름 중심으로 데이터를 셋팅하고, total기준으로 내림차순 정렬하여 상위 10개만 출력
df.groupby('역명').sum().sort_values(by='total', ascending=False)[['total']].head(10)
```

```python
top10_stations = df.groupby('역명').sum().sort_values(by='total', ascending=False)[['total']].head(10).index
top10_stations
```

```python
top10_stations_cond = df['역명'].isin(top10_stations)
top10_stations_cond
```

```python
top10 = df[top10_stations_cond]
top10

top10.노선명.unique()
```

```python
#top10.groupby('노선명').sum().sort_values(by='total', ascending=False)
top10.groupby('노선명').sum().sort_values(by='total', ascending=False).head()
```

```python
top5_line = top10.groupby('노선명').sum().sort_values(by='total', ascending=False)[['total']].head().index
top5_line
```

```python
top5_line_cond = df['노선명'].isin(top5_line)
top5_line_cond
```

```python
top = df[top10_stations_cond & top5_line_cond]
top
```

```python
top.sort_values(by='total', ascending=False)
```

```python
month_station = top.pivot_table(index='month', columns='역명', values='total')
month_station
```

```python
top5_days = may.groupby('day').sum()[['total']].sort_values(by='total', ascending=False).head().index.values
top5_days

# . 5월에 승객 수가 가장 많은 역은?
may.groupby('역명').sum()[['total']].sort_values(by='total', ascending=False)

# 승객 수가 많은 상위 10개 역 추출
top10_stations = may.groupby('역명').sum()[['total']].sort_values(by='total', ascending=False).head(10).index.values
top10_stations

# 5월 전체 피벗테이블(인기있는 날과 역만 추출)
may_table = may.pivot_table(index='역명', columns='day', values='total')
may_table.loc[top10_stations, top5_days]
```
